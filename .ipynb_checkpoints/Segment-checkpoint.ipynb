## Observations

A modified FlyingObjects class has been created to handle images from both the "images" and "gt_images" folders.

Images from the "images" folder are treated as data, while the corresponding images from the "gt_images" folder are considered ground truth.

The FlyingObjects class has been designed to convert the 3-channel RGB images from the "gt_images" folder into 2D masks based on specific class values:

Pixel values of 0,0,0 are interpreted as 0.
Pixel values of 255,0,0 are interpreted as 1.
Pixel values of 0,255,0 are interpreted as 2.
Pixel values of 0,0,255 are interpreted as 3.

This process creates masks aligned with the designated class values.

````python
class FlyingObjects(torch.utils.data.Dataset):
    """Dataset to Flying Object Dataset for the classification task.
       The label information is encoded on the filename, __extract_label will extract the label following the chosen granularity
    
    """
    def __init__(self, root,transform=None):
        super(FlyingObjects,self).__init__()
        self.root = root
        self.transform = transform

        self.images = [os.path.join(dp,f) for dp, dn, fn in os.walk(os.path.expanduser(self.root+'/image')) for f in fn if f.endswith(".png")]
        self.images.sort()
        

        self.truth = [os.path.join(dp,f) for dp, dn, fn in os.walk(os.path.expanduser(self.root+'/gt_image')) for f in fn if f.endswith(".png")]
        self.truth.sort()
    
    
    def __convert_rgb_to_class(self,image_tensor):

        classes = [
                    [0, 0, 0],      
                    [255, 0, 0],    
                    [0, 255, 0],    
                    [0, 0, 255]     
                  ]

        c, h, w = image_tensor.shape
        single_channel_tensor = torch.zeros((h, w), dtype=torch.uint8)

        for i in range(h):
            for j in range(w):
                pixel = image_tensor[:, i, j]
                for idx, in_class in enumerate(classes):
                    if torch.all(pixel == torch.tensor(in_class, dtype=torch.uint8)):
                        single_channel_tensor[i, j] = idx
                        break

        return single_channel_tensor
    
    def __convert_rgb_to_class_bydistance(self, image_tensor):
        classes = torch.tensor([
            [0, 0, 0],
            [255, 0, 0],
            [0, 255, 0],
            [0, 0, 255]
        ])

        flat_image = image_tensor.reshape(-1, 3).float() 
        classes = classes.float()  
        distances = torch.norm(flat_image.unsqueeze(1) - classes, dim=2)
        class_indices = torch.argmin(distances, dim=1)

        single_channel_tensor = class_indices.reshape(image_tensor.shape[0], image_tensor.shape[1])


        return single_channel_tensor
    
    def __normalize(self, x):
        if isinstance(x, np.ndarray):
            x = torch.from_numpy(x)
        elif isinstance(x, Image.Image):
            x = np.array(x)
            x = torch.from_numpy(x)

        x = x.float()
        min_val = x.min()
        max_val = x.max()

        if (max_val - min_val) != 0:
            normalized = ((x - min_val) / (max_val - min_val)) * 255
            thresholded = torch.where(normalized < 128, torch.tensor(0), torch.tensor(255))
        else:
            thresholded = torch.where(x < 128, torch.tensor(0), torch.tensor(255))

        return thresholded
    
    def __getitem__(self, index):
       
        x = imageio_v2.imread(self.images[index])
        x = Image.fromarray(x)
        if self.transform:
            x = self.transform(x)
        else:
            x = torch.from_numpy(x)
        x = self.__normalize(x)
        x = x.permute(2, 0, 1)
        #print(x.shape)
        
        y = imageio_v2.imread(self.truth[index])
        y = Image.fromarray(y)
        if self.transform:
            y = self.transform(y)
        else:
            y = torch.from_numpy(y)
        y = self.__normalize(y)
        #print(y.shape)
        y = self.__convert_rgb_to_class_bydistance(y)
        #print(y.shape)

        return x, y

    def __len__(self):
        return len(self.images)
````

## Create model
During training process the training loss can not only stop decreasing but also go far from a desired value. This occured due to the vanishing gradient. This can be observed as for every layer that we go backwards in the network gradient gets smaller and smaller.

Skip connections in deep architectures, skip some layer in the neural network and feeds the output of one layer as the input to the next layers. This provide an alternative path for the gradient during backpropagation. Therefore, the skip connections help to preserve the gradient.

Skip connections is also very popular for semantic segmentation where information captured in the initial layers also would like to be learned in later layers.
It has been observed that in earlier layers the learned features correspond to lower semantic information that is extracted from the input.

Therefore, for this project we will use skip connections for the auto encoder. More specifically, a U-Net. The skip connections are placed at each step of the encoder and feeding information to the decoder. The size of the input must be the same for both encoder and decoder where skip connections are applied.


![SegmentationModel](segmentation_proj.png)

````python
class SimpleModel(nn.Module):
    def __init__(self, num_classes=4, input_channels=3, **kwargs):
        super(SimpleModel, self).__init__()
        
        ## Encoder
        self.conv1 = self.Encoder(input_channels, 32, 7, 3)
        self.conv2 = self.Encoder(32, 64, 3, 1)
        
        # Decoder
        self.upconv2 = self.Decoder(64, 32, 3, 1)
        self.upconv1 = self.Decoder(32*2, num_classes, 3, 1)

    def forward(self, x):
        # Downsample
        conv1 = self.conv1(x)
        conv2 = self.conv2(conv1)
        
        # Upsample
        upconv2 = self.upconv2(conv2)
        upconv1 = self.upconv1(torch.cat([upconv2, conv1], 1))
        
        #print("hhh",upconv1.shape)

        return upconv1.float()

    def Encoder(self, in_channels, out_channels, kernel_size, padding):
        return nn.Sequential(
            torch.nn.Conv2d(in_channels, out_channels, kernel_size=kernel_size, stride=1, padding=padding),
            torch.nn.BatchNorm2d(out_channels),
            torch.nn.ReLU(),
            torch.nn.Conv2d(out_channels, out_channels, kernel_size=kernel_size, stride=1, padding=padding),
            torch.nn.BatchNorm2d(out_channels),
            torch.nn.ReLU(),
            torch.nn.MaxPool2d(kernel_size=3, stride=2, padding=1)
        )

    def Decoder(self, in_channels, out_channels, kernel_size, padding):
        return nn.Sequential(
            torch.nn.Conv2d(in_channels, out_channels, kernel_size, stride=1, padding=padding),
            torch.nn.BatchNorm2d(out_channels),
            torch.nn.ReLU(),
            torch.nn.Conv2d(out_channels, out_channels, kernel_size, stride=1, padding=padding),
            torch.nn.BatchNorm2d(out_channels),
            torch.nn.ReLU(),
            torch.nn.ConvTranspose2d(out_channels, out_channels, kernel_size=3, stride=2, padding=1, output_padding=1) 
        )
````

Do not forget to use Tensorboard. It really helps.

````python
def train_epoch(epoch, model, loader, criterion, optimizer, device):
    model.train()
    loss_list = []
    pbar = tqdm(total=len(loader), desc=f'Epoch {epoch}')
    
    for i, (data, truth) in enumerate(loader):
        data = data.float().to(device)
        truth = truth.to(device)  
        
        optimizer.zero_grad()
        logit = model.forward(data)
        loss = criterion(logit, truth)
        loss.backward()
        optimizer.step()
        
        loss_list.append(loss.item())
        pbar.update(1)
        pbar.set_postfix({'Loss': np.mean(loss_list)})
    
    pbar.close()
    return np.mean(loss_list)

def validate_epoch(epoch, model, loader, criterion, device):
    model.eval()
    loss_list = []
    correct_pixels = 0
    total_pixels = 0
    predictions = []
    true_labels = []

    intersection_list = []
    union_list = []

    pbar = tqdm(total=len(loader), desc=f'Epoch {epoch}')
    with torch.no_grad():
        for i, (data, truth) in enumerate(loader):
            data = data.float().to(device)
            truth = truth.to(device)

            logit = model.forward(data)

            loss = criterion(logit, truth)
            loss_list.append(loss.item())
            logit = torch.argmax(logit, dim=1)
            
            correct_pixels += torch.sum(logit == truth).item()
            total_pixels += truth.numel()
            
            predictions.extend(logit.view(-1).cpu().detach().numpy())
            true_labels.extend(truth.view(-1).cpu().detach().numpy())

            # Calculate IoU for each batch
            intersection = np.logical_and(truth.cpu().numpy(), logit.cpu().numpy())
            union = np.logical_or(truth.cpu().numpy(), logit.cpu().numpy())
            intersection_list.append(np.sum(intersection))
            union_list.append(np.sum(union))

            pbar.update(1)
            pbar.set_postfix({'Loss': np.mean(loss_list)})
    pbar.close()

    f1 = f1_score(true_labels, predictions, average='weighted') 
    #f1 = f1_score(true_labels, predictions, average='micro') 
    accuracy = correct_pixels / total_pixels
    jaccard = jaccard_score(np.array(true_labels), np.array(predictions), average='weighted')

    return np.mean(loss_list), f1, accuracy, jaccard
````

````python
def proper_training(model, train_loader, valid_loader, criterion, optimizer, device, total_epochs):
    epoch = 0
    train_losses = []
    val_losses = []
    val_acc = []
    val_f1 = []
    val_jacc = []
    best_val_f1 = 0
    best_val_loss = np.inf
    while epoch < total_epochs:
        train_loss = train_epoch(epoch, model, train_loader, criterion, optimizer, device)
        val_loss, f1, acc,jacc = validate_epoch(epoch, model, valid_loader, criterion, device)
        
        train_losses.append(train_loss)
        val_losses.append(val_loss)
        val_acc.append(acc)
        val_f1.append(f1)
        val_jacc.append(jacc)
        
        if best_val_loss > val_loss:
            best_val_loss = val_loss
            torch.save(model.state_dict(), 'BestModel.pt')
            print(f'Saving model at epoch: {epoch}, val loss: {val_loss}')

        print('Epoch: %d Current val loss is %.3f while the best val loss is %.3f' %
              (epoch, val_loss, best_val_loss))
        print(" ")
        
        epoch += 1
    
    return train_losses, val_losses, val_f1, val_acc, val_jacc
````

````python
train_losses, val_losses, val_f1, val_acc, val_jacc = proper_training(model, train_loader, valid_loader, criterion, optimizer, device,70)
````

````python
def matrix(loader, net, classes, label):
    y_pred = []
    y_true = []

    # iterate over test data
    for inputs, labels in loader:
        inputs, labels = inputs.float().to(device), labels.to(device)
        output = net.forward(inputs)  # Feed Network
        output = torch.argmax(output, dim=1)

        output = output.view(-1).data.cpu().numpy()  # Flatten to 1D array
        y_pred.extend(output)

        labels = labels.view(-1).data.cpu().numpy()  # Flatten to 1D array
        y_true.extend(labels)  # Save Truth

    # Build confusion matrix
    cf_matrix = confusion_matrix(y_true, y_pred)
    print(f"{Color.BLUE}{label}{Color.END}")
    print(f"{Color.BLUE}----------------------{Color.END}")
    print(" ")
    disp = ConfusionMatrixDisplay(confusion_matrix=cf_matrix, display_labels=classes)
    fig, ax = plt.subplots()
    disp.plot(ax=ax)
````

````python
def convert_class_to_rgb(single_channel_tensor):
    classes = [
        [0, 0, 0],
        [255, 0, 0],
        [0, 255, 0],
        [0, 0, 255]
    ]
    
    #print(single_channel_tensor.shape)
    h, w = single_channel_tensor.shape
    rgb_image = torch.zeros((3, h, w), dtype=torch.uint8)
    
    for i in range(h):
        for j in range(w):
            class_idx = int(single_channel_tensor[i, j].item())
            rgb_image[:, i, j] = torch.tensor(classes[class_idx], dtype=torch.uint8)
    
    return rgb_image
````

## Observation

The accuracy achieved by the model was already quite high and it appeared that the augmentations applied did not notably enhance the accuracy further. The augmentations utilized during training included:

RandomBrightnessContrast(), RandomGamma(), Blur(), Rotate(limit=5), HorizontalFlip(p=0.1), VerticalFlip(p=0.1)

It's notable to mention that augmentations impacting pixel-level characteristics were somewhat nullified by the thresholding applied after the transforms. However, augmentations like flipping and rotation seemed to introduce more training errors, potentially affecting the overall performance of the model.

````python
class FlyingObjects1(torch.utils.data.Dataset):
    """Dataset to Flying Object Dataset for the classification task.
       The label information is encoded on the filename, __extract_label will extract the label following the chosen granularity
    
    """
    def __init__(self, root,transform=None):
        super(FlyingObjects1,self).__init__()
        self.root = root
        self.transform = transform
        self.inputoutput_transforms = A.Compose([
            A.RandomBrightnessContrast(),
            A.RandomGamma(),
            A.Blur(),
            A.Rotate(limit=5),
            A.HorizontalFlip(p=0.1),
            A.VerticalFlip(p=0.1),
        ])
#         self.output_transforms = A.Compose([
#             A.ElasticTransform(),
#             A.GridDistortion(),
#         ])

        self.images = [os.path.join(dp,f) for dp, dn, fn in os.walk(os.path.expanduser(self.root+'/image')) for f in fn if f.endswith(".png")]
        self.images.sort()
        

        self.truth = [os.path.join(dp,f) for dp, dn, fn in os.walk(os.path.expanduser(self.root+'/gt_image')) for f in fn if f.endswith(".png")]
        self.truth.sort()
    
    
    def __convert_rgb_to_class(self,image_tensor):

        classes = [
                    [0, 0, 0],      
                    [255, 0, 0],    
                    [0, 255, 0],    
                    [0, 0, 255]     
                  ]

        c, h, w = image_tensor.shape
        single_channel_tensor = torch.zeros((h, w), dtype=torch.uint8)

        for i in range(h):
            for j in range(w):
                pixel = image_tensor[:, i, j]
                for idx, in_class in enumerate(classes):
                    if torch.all(pixel == torch.tensor(in_class, dtype=torch.uint8)):
                        single_channel_tensor[i, j] = idx
                        break

        return single_channel_tensor
    
    def __convert_rgb_to_class_bydistance(self, image_tensor):
        classes = torch.tensor([
            [0, 0, 0],
            [255, 0, 0],
            [0, 255, 0],
            [0, 0, 255]
        ])

        flat_image = image_tensor.reshape(-1, 3).float() 
        classes = classes.float()  
        distances = torch.norm(flat_image.unsqueeze(1) - classes, dim=2)
        class_indices = torch.argmin(distances, dim=1)

        single_channel_tensor = class_indices.reshape(image_tensor.shape[0], image_tensor.shape[1])


        return single_channel_tensor
    
    def __normalize(self, x):
        if isinstance(x, np.ndarray):
            x = torch.from_numpy(x)
        elif isinstance(x, Image.Image):
            x = np.array(x)
            x = torch.from_numpy(x)

        x = x.float()
        min_val = x.min()
        max_val = x.max()

        if (max_val - min_val) != 0:
            normalized = ((x - min_val) / (max_val - min_val)) * 255
            thresholded = torch.where(normalized < 128, torch.tensor(0), torch.tensor(255))
        else:
            thresholded = torch.where(x < 128, torch.tensor(0), torch.tensor(255))

        return thresholded
    
    def __getitem__(self, index):
        x = imageio_v2.imread(self.images[index])
        y = imageio_v2.imread(self.truth[index])

        if self.transform:
            x = Image.fromarray(x)
            y = Image.fromarray(y)

            x = self.transform(x)
            y = self.transform(y)
            x = self.inputoutput_transforms(image=np.array(x))['image']
            y = self.inputoutput_transforms(image=np.array(y))['image']
        else:
            x = torch.from_numpy(x)
            y = torch.from_numpy(y)

        # Normalize x (assuming it's an image)
        x = self.__normalize(x)

        y = self.__normalize(y)
        y = self.__convert_rgb_to_class_bydistance(y)
        
        x = x.permute(2, 0, 1)
        return x, y

    def __len__(self):
        return len(self.images)
````

````python

class SimpleModel1(nn.Module):
    def __init__(self, num_classes=4, input_channels=3, **kwargs):
        super(SimpleModel1, self).__init__()

        self.conv1 = self.Encoder(input_channels, 32, 7, 3)  
        self.conv2 = self.Encoder(32, 64, 3, 1)  

        self.upconv2 = self.Decoder(64, 64, 3, 1)  
        self.upconv3 = self.Decoder(64, 32, 3, 1) 
        self.upconv4 = self.Decoder(32, 32, 3, 1)  
        self.upconv5 = self.Decoder(64, num_classes, 3, 1)    
        

    def forward(self, x):
        #print("input",x.shape)
        conv1 = self.conv1(x)
        #print("conv1",conv1.shape)
        conv2 = self.conv2(conv1)
        #print("conv2",conv2.shape)
        
        #print("###")
        
        upconv2 = self.upconv2(conv2)
        #print("upconv2",upconv2.shape)
        upconv3 = self.upconv3(upconv2) 
        #print("upconv3",upconv3.shape)
        upconv4 = self.upconv4(upconv3)  
        #print("upconv4",upconv4.shape)
        conv1_upsampled = F.interpolate(conv1, size=upconv4.size()[2:], mode='bilinear', align_corners=True)
        upconv5 = self.upconv5(torch.cat((upconv4, conv1_upsampled), 1)) 
        #print("upconv5",upconv5.shape)
        #print("###############")

        return upconv5.float()


    def Encoder(self, in_channels, out_channels, kernel_size, padding):
        return nn.Sequential(
            torch.nn.Conv2d(in_channels, out_channels, kernel_size=kernel_size, stride=1, padding=padding),
            torch.nn.BatchNorm2d(out_channels),
            torch.nn.ReLU(),
            torch.nn.Conv2d(out_channels, out_channels, kernel_size=kernel_size, stride=1, padding=padding),
            torch.nn.BatchNorm2d(out_channels),
            torch.nn.ReLU(),
            torch.nn.MaxPool2d(kernel_size=3, stride=2, padding=1)
        )

    def Decoder(self, in_channels, out_channels, kernel_size, padding):
        return nn.Sequential(
            torch.nn.Conv2d(in_channels, out_channels, kernel_size, stride=1, padding=padding),
            torch.nn.BatchNorm2d(out_channels),
            torch.nn.ReLU(),
            torch.nn.Conv2d(out_channels, out_channels, kernel_size, stride=1, padding=padding),
            torch.nn.BatchNorm2d(out_channels),
            torch.nn.ReLU(),
            torch.nn.ConvTranspose2d(out_channels, out_channels, kernel_size=3, stride=2, padding=1, output_padding=1) 
        )
````
